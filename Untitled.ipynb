```python
cfDNA_size_profile = {}

for bed_file, dna_type in (('/mnt/orat2_stor_data3/zhouze/shared.bed','Shared DNA'),
                           ('/mnt/orat2_stor_data3/zhouze/fetal.bed','Fetal DNA')):
    with open(bed_file) as f:
        for line in f:
            chrom, start, end =line.rstrip().split()[:3]
            dna_len = int(end) - int(start)
            cfDNA_size_profile.setdefault(dna_type,{})
            cfDNA_size_profile[dna_type][dna_len] = cfDNA_size_profile[dna_type].get(dna_len,0) + 1
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12,12))

for dna_type, color in (('Shared DNA','blue'),('Fetal DNA','red')):
    x_vals, y_vals = [], []
    for key, val in sorted(cfDNA_size_profile[dna_type].items()):
        x_vals.append(key)
        y_vals.append(val*100/sum(cfDNA_size_profile[dna_type].values()))
    plt.plot(x_vals, y_vals, linewidth=2, color=color)  # plot curve

plt.ylabel('Frequency (%)', fontsize=20)
plt.xlabel('Size (bp)', fontsize=20)
plt.yticks([0, 1, 2, 3], fontsize=15)
plt.xlim(0,250)
# plt.legend(frameon=False, fontsize=15, loc='upper left')
# plt.legend()
plt.subplots_adjust(left=0.15, right=0.85, top=0.85, bottom=0.15)
plt.show()
```
